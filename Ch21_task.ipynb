[문제 2] 여러 숫자를 입력받아 그 합계를 반환하는 함수 sum_numbers를 작성하세요. 이 때, 가변 매개변수를 사용하세요.

```python
def sum_numbers(*args):
    total = 0
    for num in args:
        total += num
    return total

print("숫자를 입력하세요. (종료하려면 q 입력)")

while True:
    user_input = input("숫자 입력: ")

    if user_input.lower() == "q":   # q 입력하면 종료
        break

        try:
            num = int(user_input)
            numbers.append(num)
        except ValueError:
            print("숫자를 입력하거나 q를 입력해 종료하세요.")

# 합계 계산
result = sum_numbers(*numbers)
print("입력한 숫자들의 합계:", result)
```

[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n: int):
    for i in range(1, n + 1, 2):  # 1부터 n까지 2씩 증가 → 홀수
        yield i

n = 10

for num in odd_numbers(n):
    print(num)
```

[문제 8] 사용자로부터 3개의 숫자를 입력받아, 이를 튜플로 만든 후 최소값과 최대값을 출력하는 함수 min_max를 작성하세요.

```python
def min_max() -> None:
    numbers: Tuple[int, int, int] = tuple(
        int(input(f"{i+1}번째 숫자 입력: ")) for i in range(3)
    )

    minimum: int = min(numbers)
    maximum: int = max(numbers)

    print(f"입력한 숫자: {numbers}")
    print(f"최소값: {minimum}")
    print(f"최대값: {maximum}")

min_max()
```

[문제 9] 주어진 숫자가 짝수인지 확인하는 함수 is_even을 작성하세요. 짝수라면 True를, 아니라면 False를 반환하세요. 함수 내에서 조기 리턴을 활용하세요.

```python
def is_even(number: int) -> bool:
    if number % 2 == 0:
        return True
    else:
        return False

number = int(input("숫자 입력: "))
result = is_even(number)
print(result)
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n: int) -> int:
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print(factorial(5))
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(s1: str, s2: str) -> str:
    if len(s1) >= len(s2):
        return s1
    else:
        return s2

s1 = input("첫 번째 문자열: ")
s2 = input("두 번째 문자열: ")

result = longer_string(s1, s2)
print(f"긴 문자열: {result}")
```
